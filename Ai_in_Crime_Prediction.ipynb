```python
# DATE TIME STAMP FUNCTION
# Use .iloc for positional indexing to select the first column (index 0)
column_1 = data.iloc[:, 0]

db = pd.DataFrame({
    "year": column_1.dt.year,
    "month": column_1.dt.month,
    "day": column_1.dt.day,
    "hour": column_1.dt.hour,
    "dayofyear": column_1.dt.dayofyear,
    "week": column_1.dt.isocalendar().week,  # Use isocalendar().week
    "weekofyear": column_1.dt.isocalendar().week,  # Use isocalendar().week
    "dayofweek": column_1.dt.dayofweek,
    "weekday": column_1.dt.weekday,
    "quarter": column_1.dt.quarter,
})
```

```python
X=data1.iloc[:,[1,2,3,4,6,16,17]].values
```

```python
X
```

```python
y=data1.iloc[:,[10,11,12,13,14,15]].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split # import train_test_split from model_selection module instead of cross_validation
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=50)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn.score(X_train,y_train)
```

```python
error_rate = []
for i in range(1,140):

    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,140),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=5)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(max_depth=500, random_state=300)
```

```python
dtree.fit(X_train,y_train)
```

```python
y_pred=dtree.predict(X_test)
```

```python
dtree.score(X_test,y_test)
```

```python
dtree.score(X_train,y_train)
```

```python
y_pred
```

```python
treefeatures=dtree.feature_importances_
indices = np.argsort(treefeatures)
```

```python
treefeatures
```

```python
features = data1.iloc[:,[1,2,3,4,6,16,17]]
```

```python
plt.figure(1)
plt.title('Feature Importances')
plt.barh(range(len(indices)), treefeatures[indices], color='b', align='center')
# Get feature names based on the indices and column names of 'features'
feature_names = [features.columns[i] for i in indices]
plt.yticks(range(len(indices)), feature_names)  # Use feature_names for y-axis ticks
plt.xlabel('Relative Importance')
```
